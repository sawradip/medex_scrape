```python


generic_all_cols = set()

with open('all_herbel_generics.json', 'r', encoding='utf-8') as file:
    generics_info = json.load(file)

for generic in generics_info:
    generic_all_cols.add("generic_name")
    generic_all_cols.add("brand_count")
    generic_all_cols.add("url")
    for info_elem in generic["info"]["bn"]:
        generic_all_cols.add(info_elem["section"])
    for info_elem in generic["info"]["en"]:
        generic_all_cols.add(info_elem["section"])



with open('all_alopathic_generics_details.json', 'r', encoding='utf-8') as file:
    generics_info = json.load(file)
    
for generic in generics_info:
    generic_all_cols.add("generic_name")
    generic_all_cols.add("brand_count")
    generic_all_cols.add("url")
    for info_elem in generic["info"]["bn"]:
        generic_all_cols.add(info_elem["section"])
    for info_elem in generic["info"]["en"]:
        generic_all_cols.add(info_elem["section"])
        
len(generic_all_cols)
```

```python
global_generic_info = []


with open('all_herbel_generics.json', 'r', encoding='utf-8') as file:
    generics_info = json.load(file)

for generic in generics_info:
    generic_dict = {}
    for col in generic_all_cols:
        generic_dict[col] = ""

    generic_dict["generic_name"] = generic["generic_name"]
    generic_dict["brand_count"] = generic["brand_count"]
    generic_dict["url"] = generic["url"]
    generic_dict["type"] = "herbal"

    for info_elem in generic["info"]["bn"]:
        generic_dict[info_elem["section"]] = info_elem["text"]
    for info_elem in generic["info"]["en"]:
        generic_dict[info_elem["section"]] = info_elem["text"]

    global_generic_info.append(generic_dict)


with open('all_alopathic_generics_details.json', 'r', encoding='utf-8') as file:
    generics_info = json.load(file)
    
for generic in generics_info:
    generic_dict = {}
    for col in generic_all_cols:
        generic_dict[col] = ""

    generic_dict["generic_name"] = generic["generic_name"]
    generic_dict["brand_count"] = generic["brand_count"]
    generic_dict["url"] = generic["url"]
    generic_dict["type"] = "allopathic"

    for info_elem in generic["info"]["bn"]:
        generic_dict[info_elem["section"]] = info_elem["text"]
    for info_elem in generic["info"]["en"]:
        generic_dict[info_elem["section"]] = info_elem["text"]

    global_generic_info.append(generic_dict)

```
